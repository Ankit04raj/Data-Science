# Oprater Precendence

```python
# Operator Precedence
# Python follows PEMDAS (Parentheses, Exponents, Multiplication/Division, Addition/
# Subtraction). The order of operations in Python is:
# 1. Parentheses () – Highest precedence, operations inside parentheses are
# evaluated first.
# 2. Exponents ** – Power calculations (e.g., 2 ** 3 → 8).
# 3. Multiplication * , Division / , Floor Division // , Modulus % – Evaluated
# from left to right.
# 4. Addition + , Subtraction - – Evaluated from left to right.

# Example:
result = 10 + 2 * 3
print(result)
# Multiplication happens first: 10 + (2 * 3) = 16
result = (10 + 2) * 3
print(result)# Parentheses first: (10 + 2) * 3 = 36
result = 2 ** 3 ** 2
print(result)# Right-to-left exponentiation: 2 ** (3 ** 2)
```

# Recursion

```python
def countdown(n):
    if n == 0:
        print("currapted")
    else:
        print(n)
        countdown(n-1)
countdown(5)
```

```python
def factorial(n):
    if n ==0 | n ==1:
        return 1
    return n * factorial(n-1)
factorial(5)
```

# Match Case in Python

```python
code = int(input("enter the code :"))
match code:
    case 200:
        print("response Ok")
    case 400:
        print("bad response")
    case 404:
        print("found error")
print("End of the program")
```

```python
# code = int(input("enter the code :"))
def http_status(code):
    match code:
        case 200:
            return "OK"
        case 400:
            return "Bad Request"
        case 404:
            return "Not Found"
        case 500:
            return "Internal server error"
        case _:
            return "Unlock Status"

http_status(200)
http_status(404)
```

```python
# # serialization & deserialization  custom object
# class Person:
#     def __init__(self,fname,lname,age,gender):
#         self.fname = fname
#         self.lname = lname
#         self.age= age
#         self.gender = gender


#     #format to printed in
#     # -> Ankit raj age -> 21  gender -> male
```

```python
# serialization & deserialization  custom object
class Person:
    def __init__(self,fname,lname,age,gender):
        self.fname = fname
        self.lname = lname
        self.age= age
        self.gender = gender


    #format to printed in
    # -> Ankit raj age -> 21  gender -> male

import json

def show_object(person):
    if isinstance(person,Person):
        return "{} {} age -> {} gender -> {}".format(person.fname,person.lname,person.age,person.gender)

person= Person("Ankit","raj",21,"male")
with open("demo.json","w") as f:
    json.dump(person,f,default=show_object ,indent =4)
```

```python
#AS A DICTIONARY
import json

def show_object(person):
    if isinstance(person,Person):
        return {
            "name":person.fname + " " +person.lname ,"age":person.age,"gender":person.gender
        }

person= Person("Ankit","raj",21,"male")
with open("demo.json","w") as f:
    json.dump(person,f,default=show_object ,indent =4)
```

```python
#deserialization
import json
with open("demo.json","r") as f:
    print(json.load(f))
```

```python
class Bank:
    def __init__(self,balance):
        self.balance = balance

    def withdraw(self,amount):
        if amount < 0:
            raise Exception("amount can't bve -ve")
        if self.balance < amount:
            raise Exception("you don't have sufficient balancre")
        self.balance -= amount

obj = Bank(10000)
try:
    obj.withdraw(15000)
except Exception as e:
    print(e)
else:
    print(obj.balance)
```

# Scope and LEGB Rule
A scope is a textual region of a Python program where a namespace is directly accessible.The interpreter searches for a name from the inside out, looking in the local, enclosing, global, and finally the built-in scope. If the interpreter doesn’t find the name in any of these locations, then Python raises a NameError exception

```python
# "nonlocal" keyword
# change outer function by inner

def outer():
    a = 1
    def inner():
        nonlocal a  # change outer function
        a += 1
        print("inner",a)
    inner()
    print("outer",a)

outer()
print("main function")
```

# Decorators
A decorator in python is a function that receives another function as input and adds some functionality(decoration) to and it and returns it.

This can happen only because python functions are 1st class citizens.

There are 2 types of decorators available in python

    ** Built in decorators like @staticmethod, @classmethod, @abstractmethod and @property etc
    ** User defined decorators that we programmers can create according to our needs


```python
#python are first class function
def modify(func,num):
    return func(num)

def square(num):
    return num **2


modify(square,2)
```

```python
#simple exzample
def my_decorator(func):
    def wrapper():
        print("**************************************")
        func()
        print("**************************************")
    return wrapper

def hello():
    print("hello")
def display():
    print("hi Ankit")

x = my_decorator(hello)
x()
a = my_decorator(display)
a()
```

```python
#better syntaax
#simple example
def my_decorator(func):
    def wrapper():
        print("**************************************")
        func()
        print("**************************************")
    return wrapper

@my_decorator
def hello():
    print("hello")
@my_decorator
def display():
    print("hi Ankit")

hello()
display()
```

```python
# meaningful decorator
import time

def timer(func):
    def wrapper(*args):
        start = time.time()
        func(*args)
        print("time taken by",func.__name__,time.time()-start,"secs")
    return wrapper

@timer
def hello():
    print("hello world!")
    time.sleep(2)

@timer
def square(num):
    time.sleep(1)
    print(num**2)

@timer
def power(a,b):
    print(a**b)

hello()
square(3)
power(3,2)
```

```python
# decorater with arguments
# @checkdt(int)   # not define u have to define function as checkdt
def square(num):
    print(num**2)
```

```python
def sanity_check(data_type):
    def outer_wrapper(func):
        def inner_wrapper(*args):
            if type(*args) == data_type:
                func(*args)
            else:
                raise TypeError("this dattype is invalid")
        return inner_wrapper
    return outer_wrapper

@sanity_check(int)
def square(num):
    print(num**2)

@sanity_check(str)
def greet(name):
    print("hello",name)

square(2)
greet("Ankit")
```
