```python
import numpy as np
import pandas as pd 
```

```python
df = pd.read_csv("/home/ankit/Data_Science/Machine Learning/placement.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.iloc[:,1:]
```

```python
df.head()
```

```python
# Steps
# 0. Preprocess + EDA + Feature Section
# 1. Extract input and out put Cols
# 2. Scale the value
# 3. train test split
# 4. train the model
# 5. Evaluate the model/model selection
# 6. Deploy the model
```

```python
import matplotlib.pyplot as plt
```

```python
# step : 0- preprocess + EDA + Feature Section
# cgpa  $ iq 
plt.scatter(df["cgpa"],df["iq"])
```

```python
# got palaced 
plt.scatter(df["cgpa"],df["iq"],c= df["placement"])
```

```python
# step 1:-Extract input and out put Cols
# we want all the input and output
x = df.iloc[:,:2]  # input (cgpa +iq)
y = df.iloc[:,-1]  # output (placement)
```

```python
x
```

```python
y.shape
```

```python
# step 3:-train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

```python
# step 2. Scale the value
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
# step 4. train the model
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
# model Training
clf.fit(x_train,y_train)
```

```python
# step 5. Evaluate the model/model selection
y_pred = clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)   
```

```python
# how to plot the Decision boundary
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x_train,y_train.values,clf=clf ,legend =2)
```

```python
# step 6. Deploy the model

import pickle
```

```python
# with open("model.pkl","wb") as f:
#     pickle.dump(clf,f)
```

```python
pickle.dump(clf,open("model.pkl","wb"))
```

## Convertors

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv("ipl-matches.csv",converters={"Team1":rename}).head(2)
```
