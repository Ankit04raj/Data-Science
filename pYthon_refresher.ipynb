# Oprater Precendence

```python
# Operator Precedence
# Python follows PEMDAS (Parentheses, Exponents, Multiplication/Division, Addition/
# Subtraction). The order of operations in Python is:
# 1. Parentheses () – Highest precedence, operations inside parentheses are
# evaluated first.
# 2. Exponents ** – Power calculations (e.g., 2 ** 3 → 8).
# 3. Multiplication * , Division / , Floor Division // , Modulus % – Evaluated
# from left to right.
# 4. Addition + , Subtraction - – Evaluated from left to right.

# Example:
result = 10 + 2 * 3
print(result)
# Multiplication happens first: 10 + (2 * 3) = 16
result = (10 + 2) * 3
print(result)# Parentheses first: (10 + 2) * 3 = 36
result = 2 ** 3 ** 2
print(result)# Right-to-left exponentiation: 2 ** (3 ** 2)
```

# Match Case in Python

```python
code = int(input("enter the code :"))
match code:
    case 200:
        print("response Ok")
    case 400:
        print("bad response")
    case 404:
        print("found error")
print("End of the program")
```

```python
# code = int(input("enter the code :"))
def http_status(code):
    match code:
        case 200:
            return "OK"
        case 400:
            return "Bad Request"
        case 404:
            return "Not Found"
        case 500:
            return "Internal server error"
        case _:
            return "Unlock Status"

http_status(200)
http_status(404)
```

```python
# # serialization & deserialization  custom object
# class Person:
#     def __init__(self,fname,lname,age,gender):
#         self.fname = fname
#         self.lname = lname
#         self.age= age
#         self.gender = gender


#     #format to printed in
#     # -> Ankit raj age -> 21  gender -> male
```

```python
# serialization & deserialization  custom object
class Person:
    def __init__(self,fname,lname,age,gender):
        self.fname = fname
        self.lname = lname
        self.age= age
        self.gender = gender


    #format to printed in
    # -> Ankit raj age -> 21  gender -> male

import json

def show_object(person):
    if isinstance(person,Person):
        return "{} {} age -> {} gender -> {}".format(person.fname,person.lname,person.age,person.gender)

person= Person("Ankit","raj",21,"male")
with open("demo.json","w") as f:
    json.dump(person,f,default=show_object ,indent =4)
```

```python
#AS A DICTIONARY
import json

def show_object(person):
    if isinstance(person,Person):
        return {
            "name":person.fname + " " +person.lname ,"age":person.age,"gender":person.gender
        }

person= Person("Ankit","raj",21,"male")
with open("demo.json","w") as f:
    json.dump(person,f,default=show_object ,indent =4)
```

# Decorators
A decorator in python is a function that receives another function as input and adds some functionality(decoration) to and it and returns it.

This can happen only because python functions are 1st class citizens.

There are 2 types of decorators available in python

    ** Built in decorators like @staticmethod, @classmethod, @abstractmethod and @property etc
    ** User defined decorators that we programmers can create according to our needs


```python
#python are first class function
def modify(func,num):
    return func(num)

def square(num):
    return num **2


modify(square,2)
```

```python
#simple exzample
def my_decorator(func):
    def wrapper():
        print("**************************************")
        func()
        print("**************************************")
    return wrapper

def hello():
    print("hello")
def display():
    print("hi Ankit")

x = my_decorator(hello)
x()
a = my_decorator(display)
a()
```

```python
#better syntaax
#simple example
def my_decorator(func):
    def wrapper():
        print("**************************************")
        func()
        print("**************************************")
    return wrapper

@my_decorator
def hello():
    print("hello")
@my_decorator
def display():
    print("hi Ankit")

hello()
display()
```

```python
# meaningful decorator
import time

def timer(func):
    def wrapper(*args):
        start = time.time()
        func(*args)
        print("time taken by",func.__name__,time.time()-start,"secs")
    return wrapper

@timer
def hello():
    print("hello world!")
    time.sleep(2)

@timer
def square(num):
    time.sleep(1)
    print(num**2)

@timer
def power(a,b):
    print(a**b)

hello()
square(3)
power(3,2)
```

```python
# decorater with arguments
# @checkdt(int)   # not define u have to define function as checkdt
def square(num):
    print(num**2)
```

```python
def sanity_check(data_type):
    def outer_wrapper(func):
        def inner_wrapper(*args):
            if type(*args) == data_type:
                func(*args)
            else:
                raise TypeError("this dattype is invalid")
        return inner_wrapper
    return outer_wrapper

@sanity_check(int)
def square(num):
    print(num**2)

@sanity_check(str)
def greet(name):
    print("hello",name)

square(2)
greet("Ankit")
```

### CREATING A NEW DATA TYPE 

```python
#NEW DATA TYPE IS fraction DATA TYPE
class fraction:
    # parameterized constructor
    def __init__(self,x,y):
        self.num =x      # numirator
        self.den = y     # denometer

    # 2nd magic method
    def __str__(self):  
        return "{}/{}".format(self.num,self.den)

    #3rd magic method
    def __add__(self,other):
        new_num= self.num*other.den + other.num *self.den
        new_den = self.den * other.den

        return "{}/{}" .format(new_num,new_den)
        
    # 4th magic method
    def __sub__(self,other):
        new_num = self.num*other.den -other.num*self.den
        new_den = self.den * other.den
        return "{}/{}".format(new_num,new_den)

    # 5th magic method
    def __mul__(self,other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return "{}/{}".format(new_num,new_den)

    # 6th magic method
    def __truediv__(self,other):
        new_num = self.num*other.den
        new_den = self.den*other.num
        return "{}/{}".format(new_num,new_den)

    # 7th magic number
    # convert_to_decimal
    def convert_to_decimal(self):
        return self.num/self.den
```

### Write OOP class to hande the following scenarios
#### 1> A user can create and view 2D coordinates
#### 2> A user can find out the distance betweem 2 coordinates
#### 3> A user can find the disstance of cordinate from ogin
#### 4> A user can check if a point lies on a group line
##### 5> A user can find the distance between a given 2D point and a given line

```python
class Point:

    def __init__(self,x,y):
        self.x_cod = x
        self.y_cod = y
# 1> A user can create and view 2D coordinates
    def __str__(self):
        return "<{},{}>".format(self.x_cod,self.y_cod)

# 2> A user can find out the distance between 2 coordinates
    def eculidian_distance(self,other):
        return ((self.x_cod - other.x_cod)**2 + (self.y_cod - other.y_cod)**2)**0.5

# 3> A user can find the disstance of cordinate from ogin
    def distance_from_origin(self):
        # return (self.x_cod**2 + self.y_cod**2)**0.5
        return self.eculidian_distance(Point(0,0))

# 4> A user can check if a point lies on a group line
    
        
class Line:
    def __init__(self,A,B,C):
        self.A = A
        self.B = B
        self.C = C

    def __str__(self):
        return "{}x + {}y + {} = 0".format(self.A,self.B,self.C)

    def point_on_line(line,point):
        if line.A*point.x_cod + line.B*point.y_cod +line.C ==0:
            return "point lies on the line"
        else:
            return "point doesnot lies on the line"


# 5> A user can find the distance between a given 2D point and a given line

    def shortest_distance(line,point):
        return abs(line.A*point.x_cod + line.B*point.y_cod + line.C)/(line.A**2 + line.B**2)**0.5        
        
```

```python
# non private attribute
# child cannot accesss private member of the class
# example 3
class Phone:
    def __init__(self,price,brand,camera):   
        print("inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def show(self):
        print(self.__price)

class Smartphone(Phone):
    def check(self):
        print(self.__price)

s = Smartphone(200000,"Apple",64)
print(s.brand) #  its doesnot shows error


# s.check()  #  show error bcz its parent class private attribute
```

```python
# example 4
# we can call private inytity by using getter method

class Parent:

    def __init__(self,num):
        self.__num = num

    def get_num(self):
        return self.__num

class Child(Parent):

    def show(self):
        print("this is in child class")

son = Child(100)
print(son.get_num())
son.show()
```

#### METHOD OVERRIDING

```python
# MTHOD OVERRIDING
# IF CHILD CLASS AND PARENT CLASS BOTH HAVE SAME 
# TYPES OF METHOD THEN CHILD CLASS METHOD IS CALLED

class Phone:
    def __init__(self,price,brand,camera):   
        print("inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("buying a phone")

class Smartphone(Phone):
    def buy(self):
        print("buying a smartphoe")

s = Smartphone(20000,"apple",64)
s.buy()
```

#### SUPER KEYWORD

```python
#  supersuper :- keyword  used in side the  child class
class Phone:
    def __init__(self,price,brand,camera):   
        print("inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("buying a phone")

class Smartphone(Phone):
    def buy(self):
        print("buying a smartphoe")
        # SYNTEX TO CALL PARENT "BUY METHODS"
        super().buy()

s = Smartphone(20000,"apple",64)
s.buy()
# super().buy()    # its gives error bcz it is outside the class
```

```python
# super -> constructor 
# its calls previous Phone(parent) class

class Smartphone(Phone):
    def __init__(self,price,brand,camera,os,ram):
        super().__init__(price,brand,camera)
        self.os = os
        self.ram = ram
        print("inside smart phone constructor")

s = Smartphone(20000,"samsung",12,"android",2)

print(s.os)
print(s.brand)
```

#### SUMMARY OF INHERITANCE
##### a> a class can inherit from another class
##### b> inheritance improves code reuse
##### c> constructor ,attributes,method get inherited to the child class
##### d> the parent has no access to the child class
##### e> private properties of parent are not accessible directly in child class
##### f> child class can override the attributes or method.this is called method overriding
##### g> super() is an inbuilt function which is used to invoke the parent class method and constructor

### Types of INHERITANCE
#### 1> single inheritance
#### 2> Multilevel inheritance
#### 3> Hierarchial inheritance
#### 4> multiple inheritance (diamond problem)
#### 5> Hybrid inheritance

```python
# 1> single inheritance
class Phone:
    def __init__(self,price,brand,camera):   
        print("inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("buying a phone")

class Smartphone(Phone):
    pass

s = Smartphone(20000,"apple",64)
s.buy()
```

```python
# 2> multilevel Inheritance
class Product():
    def review(self):
        print("product customer review")

class Phone(Product):
    def __init__(self,price,brand,camera):   
        print("inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("buying a phone")

class Smartphone(Phone):
    pass

s = Smartphone(20000,"apple",64)
s.buy()
s.review()
```

```python
#3>  Hierarchial
class Phone(Product):
    def __init__(self,price,brand,camera):   
        print("inside phone constructor")
        self.price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("buying a phone")

class Smartphone(Phone):
    pass

class Featurephone(Phone):
    pass


s = Smartphone(20000,"apple",64)
s.buy()

Featurephone(10000,"lava",12).buy()
```

```python
# 4> Multiple (diamond problem) 
class Phone:
    def __init__(self,price,brand,camera):
        print("inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("buying a phone")

class Product:
    def review(self):
        print("Customer review")

class Smartphone(Phone,Product):
    pass 

s = Smartphone(20000,"apple",64)
s.buy()
s.review()
```

```python
# example 2
class Phone:
    def __init__(self,price,brand,camera):
        print("inside phone constructor")
        self.__price = price
        self.brand = brand
        self.camera = camera

    def buy(self):
        print("buying a phone")

class Product:
    def review(self):
        print("Customer review")
# MRO (METHOD RESOLUTION ORDER) 
class Smartphone(Phone,Product):  #  the first parent
    pass 

s = Smartphone(20000,"apple",64)
s.buy()   #  will  call here (the 1st parent)
```

```python
# example:
class A:
    def m1(self):
        return 20

class B(A):
    def m1(self):
        return 30

    def m2(self):
        return 40

class C(B):
    def m2(self):
        return 20

obj1 = A()
obj2 = B()
obj3 = C()
print(obj1.m1() + obj3.m1() + obj3.m2())
```

## 4> POLYMORPHISM :- 
#### same things act like diff - diff according to situation

###  **method overriding :-
#### in senario of inheritance parent and child has same method naame and child obj calls that method then the execution in from child class  

### **method overloading:-  (it DOESN'T WORK IN PYTHON)
#### in a single class there are two same methods of same anme but there behavior are diff depending on the input

### **operator overloading :- 
#### same operater but depending on diff input its behaviour is different

```python
#method overrloading:-
# in a single class there are two same methods of same anme but there behavior are diff depending
# whole reason of method overriding is the code will be cleaner

# method overloading doesnot work in python
  #  but there is a smart way to implement method overloading

# example
class Shape:
    def area(self,a,b=0):
        if b == 0:
            return 3.14*a*a
        else:
            return a*b

s = Shape()
print(s.area(2))
print(s.area(3,4))
```

```python
# operator overloading :-
# same operater but depending on diff input its behaviour is different

print("hello" + "world")
print(4+5)
print([1,4,3]+[5,7])
```

##  ABSTRACTION :-  hide detail
#### by using abstraction we can force to lower level class by heigher level class to do some stuff

```python
from abc import ABC,abstractmethod
class BankApp(ABC):

    def database(self):
        print("connected to database")

    @abstractmethod
    def security(self):
        pass
    
    
```

```python
class MobileApp(BankApp):

    def mobile_login(self):
        print("login into mobile")

    def security(self):  # we have to use security from higher level class
        print("mobile security")
```

```python
mob = MobileApp()
```

```python
mob.mobile_login()
```
