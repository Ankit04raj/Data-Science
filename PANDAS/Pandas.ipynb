```python
# Data frame
points_df = pd.DataFrame({
    "1st point":[(3,4),(-6,5),(0,0),(-10,1),(4,5)],
    "2nd point":[(-3,4),(0,0),(2,2),(10,10),(1,1)]
})
points_df
```

```python
def euclidean(row):
    pt_A = row["1st point"]
    pt_B = row["2nd point"]

    return ((pt_A[0]-pt_B[0])**2 + (pt_A[1]-pt_B[1])**2)**0.5
```

```python
points_df["distance"] = points_df.apply(euclidean,axis=1)  # axis = 1 is row
points_df
```

```python
movies = pd.read_csv("imdb-top-1000.csv")
```

```python
movies.head(1)
```

```python
movies.shape
```

```python
movies.head()
```

```python
genres = movies.groupby("Genre")
genres
```

```python
# Applying builtin aggregation function on groupby objects
genres.min().sample(2)
```

```python
genres.sum().head(2)
```

```python
# find the top 3 genres by total earning
gross = movies.groupby("Genre")
gross
```

```python
gross.sum()["Gross"].sort_values(ascending = False).head(3)
```

```python
# find the top 3 genres by total earning
#  2 ND METHOD 
movies.groupby("Genre")["Gross"].sum().sort_values(ascending = False).head(3)
```

```python
 # find the genre with highest avg IMDB rating
```

```python
movies.groupby("Genre")["IMDB_Rating"].mean().sort_values(ascending = False).head(1)
```

```python
# find the director with most popularity
#  take no_of_votes as  reference of popularity

movies.groupby("Director")["No_of_Votes"].sum().sort_values(ascending = False).head(1)
```

```python
# find thye highest rated movie of each genre
movies.groupby("Genre")["IMDB_Rating"].idxmax().sort_values(ascending = False)
```
