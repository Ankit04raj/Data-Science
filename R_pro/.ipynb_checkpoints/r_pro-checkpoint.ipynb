```r
# 5> armstrong number
num <- as.integer(readline("Enter a number: "))
n_num <- num
len <- nchar(num)

count <-0
while(n_num > 0){
    digit <- n_num%%10
    count <- count +digit^len
    n_num <- n_num%/%10
}
if (count == num){
    cat("Armstrong")
}else{
    cat("not an armstrong")
}
```

```r
# 4 find variance and standurd deviation
data <- c(5,15,25,10,20)
n <- length(data)

sum <- 0
for (i in 1:n){
    sum <- sum + data[i]
} 
mean_val <- sum/n


var_sum <- 0 
for (i in 1:n){
    new <- (data[i]- mean_val)
    var_sum <- var_sum +(new*new)
}

variance <- var_sum/(n-1)
std_dev <- sqrt(variance)
cat("variance",variance)
cat("std_dev", std_dev)
```

```r
# inverse of a matrix
A <- matrix(c(4,6,7,4),nrow = 2 ,byrow = TRUE)

a<- A[1,1] 
b<- A[1,2] 
c<- A[2,1] 
d<- A[2,2]

determinant <- a*d - b*c

if(determinant == 0){
    cat("inverse does not exist\n")

}else{
    invA <- (1/determinant) * matrix(c(d,-b,-c,a),nrow=2,byrow=TRUE)
    print(invA)
}
```

```r
# Inheritance example in R

# Parent class
person <- list(name = "Ankit", age = 21)
class(person) <- "Person"

cat("Person Details\n")
print(person)

# Child class (inherits from Person)
student <- list(name = "Arijit", age = 20, marks = 92)
class(student) <- c("Student", "Person")  # Inheritance

cat("\nStudent Details\n")
print(student)

cat("\nClass of student object:\n")
print(class(student))
```

```r
# Encapsulation 

create_account <- function(initial_balance) {
  
  balance <- initial_balance   
  
  list(
    # method to deposit money
    deposit = function(amount) {
      balance <<- balance + amount
      cat("Deposited", amount, "\n")
    },
    
    # method to withdraw money
    withdraw = function(amount) {
      if (amount > balance) {
        cat("Insufficient balance\n")
      } else {
        balance <<- balance - amount
        cat("Withdrawn", amount, "\n")
      }
    },
    
    # method to check balance
    get_balance = function() {
      cat("Current Balance:", balance, "\n")
    }
  )
}

```

```r
# use switch case
# 1> account_balance
# 2> how much money deposited
# 3> how many money withdraw
# 4> how many final get_balnce

account <- create_account(5000)
account$get_balance()
account$deposit(2000)
account$withdraw(1000)
account$get_balance()
```

```r
library(readxl)
library(writexl)

file_path <- "StudentR.xlsx"
output_folder <- "Attendance_Records"

if (!dir.exists(output_folder)) {
  dir.create(output_folder)
}

date_input <- readline("Enter date (dd.mm.yyyy): ")
date_obj <- as.Date(date_input, format = "%d.%m.%Y")
day_of_week <- weekdays(date_obj)

data <- read_excel(file_path)

attendance <- c()

cat("\nEnter Attendance (P = Present, A = Absent)\n")

for (i in 1:nrow(data)) {
  repeat {
    status <- readline(
      paste("Student:", data$Name[i], "(P/A): ")
    )
    status <- toupper(status)

    if (status %in% c("P", "A")) {
      attendance[i] <- status
      break
    } else {
      cat(" Invalid input! Enter P or A only\n")
    }
  }
}

data$Attendance <- attendance
data$Date <- date_input
data$Day <- day_of_week

output_file <- paste0(
  output_folder, "/Attendance_", gsub("\\.", "-", date_input), ".xlsx"
)

write_xlsx(data, output_file)

cat("\nAttendance saved successfully at:\n", output_file)
```

```r
library(readxl)

data_today <- read_excel(output_file)

attendance_count <- table(data_today$Attendance)

print(attendance_count)

# Bar plot
par(bg="white")
barplot(
  attendance_count,
  main = paste("Attendance Summary -", date_input),
  xlab = "Attendance",
  ylab = "Number of Students",
  col = c("skyblue", "orange")
)
```

```r
library(readxl)

files <- list.files(
  "Attendance_Records",
  pattern = "\\.xlsx$",
  full.names = TRUE
)

all_attendance <- do.call(rbind, lapply(files, read_excel))

all_attendance$Present <- ifelse(
  all_attendance$Attendance == "P", 1, 0
)

attendance_numeric <- aggregate(
  Present ~ Name,
  data = all_attendance,
  sum
)

colnames(attendance_numeric) <- c("name", "total_present")

print(attendance_numeric)
```

```r
par(bg="white")
barplot(
  attendance_numeric$total_present,
  names.arg = attendance_numeric$name,
  main = "Total Present Days per Student",
  xlab = "Students",
  ylab = "Number of Present Days",
  col = "orange",
  las = 2
)
```

```r
# Convert Absent to numeric
all_attendance$Absent <- ifelse(
  all_attendance$Attendance == "A", 1, 0
)

attendance_absent <- aggregate(
  Absent ~ Name,
  data = all_attendance,
  sum
)

colnames(attendance_absent) <- c("name", "total_absent")

print(attendance_absent)
```

```r
par(bg="white")
barplot(
  attendance_absent$total_absent,
  names.arg = attendance_absent$name,
  main = "Total Absent Days per Student",
  xlab = "Students",
  ylab = "Number of Absent Days",
  col = "red",
  las = 2
)
```

```r
# WAP TO CHECK A NUMBER IS KRISHNAMURTHY OR NOT
# THORA HELP ;LYA HU 
fact <- function(x) {
  f <- 1
  for (i in 1:x) {
    f <- f * i
  }
  return(f)
}
```

```r
num <- as.integer(readline("Enter a number: "))
temp <- num
sum <- 0

while (temp > 0) {
  digit <- temp %% 10
  sum <- sum + fact(digit)
  temp <- temp %/% 10
}

if (sum == num) {
  cat(num, "HMM  YE TO HAIN Krishnamurthy number")
} else {
  cat(num, "YE NAHI HAIN Krishnamurthy number")
}
```

```r
# WAP TO PRINT PRIME NUMBER BETWEEN 1 TO N
n <- as.integer(readline("Enter the value of N: "))

for (X in 2:n) {
  count <- 0
  
  for (i in 1:X) {
    if (X %% i == 0) {
      count <- count + 1
    }
  }
  
  if (count == 2) {
    cat(X, " ")
  }
}
```
